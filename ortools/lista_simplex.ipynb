```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver('GLOP')
```

```python
xa = solver.NumVar(0, solver.infinity(), 'xa')
xb = solver.NumVar(0, solver.infinity(), 'xb')
xc = solver.NumVar(0, solver.infinity(), 'xc')

print('Number of variables =', solver.NumVariables())
```

```python
solver.Add(0.2*xa + 0.1*xb + 0.3*xc <= 500)
solver.Add(0.06*xa + 0.04*xb + 0.08*xc <= 168)

print('Number of constraints =', solver.NumConstraints())
```

```python
solver.Maximize(0.3*xa + 0.25*xb + 0.1*xc)
```

```python
status = solver.Solve()
```

```python
if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    print('xa =', xa.solution_value())
    print('xb =', xb.solution_value())
    print('xc =', xc.solution_value())
else:
    print('The problem does not have an optimal solution.')
```

```python
print(solver.ExportModelAsLpFormat(False))
```
